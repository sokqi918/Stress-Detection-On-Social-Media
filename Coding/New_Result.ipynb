```python
import pandas as pd
df = pd.read_csv('/content/comparison_table.csv')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = df[df['Feature_Extraction'] == 'BoW']
data

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a grouped bar chart based on Models
plt.figure(figsize=(12, 8))

# Melt the DataFrame to have a single column for metrics
melted_data = pd.melt(data, id_vars=['Models'], value_vars=['Accuracy', 'Precision', 'Recalls', 'F1-Score'])

# Plot the grouped bar chart
sns.barplot(x='Models', y='value', hue='variable', data=melted_data, ci=None)

plt.title('Performance Metrics for Different Models (with BoW)')
plt.ylabel('Metrics (%)')
plt.legend(title='Metric')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame with the performance metrics
data = df[df['Feature_Extraction'] == 'BoW']

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a grouped bar chart based on Models
plt.figure(figsize=(8, 7))

# Melt the DataFrame to have a single column for the F1-Score metric
melted_data = pd.melt(data, id_vars=['Models'], value_vars=['F1-Score'])

# Define a single color for all bars (bright blue)
bar_color = '#2907ca'

# Adjust the bar width
bar_width = 0.6  # Adjust as needed

# Plot the grouped bar chart with a single color and adjusted bar width
ax = sns.barplot(x='Models', y='value', data=melted_data, ci=None, color=bar_color, dodge=True, linewidth=0.5, edgecolor='black', width=bar_width)

# Display values on the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}%', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 10),
                textcoords='offset points')

plt.title('F1-Score for Different Models (with BoW)')
plt.ylabel('Metrics (%)')
plt.show()
```

```python
import pandas as pd

data = {
    'Models': ['SVM with BoW', 'LR with BoW', 'NB with BoW', 'DT with BoW', 'RF with BoW',
               'SVM with TF-IDF', 'LR with TF-IDF', 'NB with TF-IDF', 'DT with TF-IDF', 'RF with TF-IDF',
               'SVM with Word2Vec', 'LR with Word2Vec', 'NB with Word2Vec', 'DT with Word2Vec', 'RF with Word2Vec',
               'LSTM with GloVe', 'CNN with GloVe'],
    'F1-Score': [75.61, 77.67, 79.76, 62.74, 76.76,
                 77.45, 78.26, 78.42, 57.37, 77.38,
                 63.17, 68.19, 63.27, 56.24, 65.7,
                 72.78, 76.44]
}

df2 = pd.DataFrame(data)

print(df2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you already have the DataFrame df

# Sort the DataFrame by 'F1-Score' in descending order
df_sorted = df2.sort_values(by='F1-Score', ascending=False)

# Plotting the bar chart with color set to #008000 (green)
plt.figure(figsize=(10, 6))
plt.bar(df_sorted['Models'], df_sorted['F1-Score'], color='#008000')
plt.xlabel('Models')
plt.ylabel('F1-Score')
plt.title('F1-Score for Different Models')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display the F1-Score values on top of the bars
for i, value in enumerate(df_sorted['F1-Score']):
    plt.text(i, value + 0.5, f'{value:.2f}', ha='center', va='bottom', fontsize=8)

plt.show()
```
